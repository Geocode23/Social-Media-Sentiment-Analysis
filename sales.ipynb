```python
import pandas as pd
```

```python
pd=pd.read_csv('train.csv')
```

```python
clean_pd=pd.drop_duplicates()   
```

```python
#group sales based on item category
category_sales=(
    clean_pd.groupby('Category')['Sales']
    .sum()
    .reset_index()
    .sort_values(by='Sales', ascending=False)
)
```

```python
category_sales
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.bar(category_sales['Category'], category_sales['Sales'], color='skyblue')
plt.xlabel('Item Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
```

```python
clean_pd['Order Date'] = pd.to_datetime(
    clean_pd['Order Date'],
    errors='coerce'
)
```

```python
clean_pd['month'] = clean_pd['Order Date'].dt.month
```

```python
#aggrigate sales by month
monthly_sales=(
    clean_pd.groupby('month')['Sales']
    .sum()
    .reset_index()
    .sort_values(by='month')
)
```

```python
monthly_sales = (
    clean_pd.groupby('month')['Sales']
    .sum()
    .reset_index()
    .sort_values(by='month')
)

plt.figure(figsize=(10,6))
plt.plot(monthly_sales['month'], monthly_sales['Sales'], marker='o', color='orange')
plt.xlabel('Month') 
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')
plt.xticks(range(1, 13))
plt.grid()
plt.show()
```

```python
#monthly avarage sales
monthly_sales['Avg Monthly Sales']=(clean_pd.groupby('month')['Sales'].mean().values)
```

```python
#month over month growth
monthly_sales['MoM Growth (%)'] = (monthly_sales['Sales'].pct_change() * 100)
```
